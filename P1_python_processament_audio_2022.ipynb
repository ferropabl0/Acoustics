## 1.4 Python per al processament de senyal

Anem a veure una mica com aplicar aquests coneixements bàsics de Python per al processament de senyals d'àudio.

```python
fs = 16000.0 # definim una freqüència de mostreig

tt = np.arange(0, 1, 1/fs) # vector temporal de 0 a 1 segons en passos de 1/fs unitats

y = 2 * np.sin(2 * np.pi * 20 * tt) # creem una sinusoïde de freqüència 20 Hz i amplitud 2

# grafiquem
plt.figure(figsize=(10,5)) # definim la mida que volem per la figura 
plt.plot(tt, y, 'r-')
plt.title("Sinus", fontsize=15)
plt.xlabel("Angle [rad]"), plt.ylabel("Amplitud")
plt.show()
```

```python
# codi de la funció aquí
def diferents(llista):

    new_list = []

    for element in llista:

        if element in new_list:
            return False
        new_list.append(element)

    return True

# crideu aquí la funció
true_list = np.arange(9)
false_list = np.zeros(3)
print('[0,8]:', diferents(true_list))
print('[0,0,0]:', diferents(false_list))
```

## Exercici 1: sinusoïdes i els seus paràmetres (2 punts)

De teoria sabem que l'equació de desplaçament d'un sistema massa-molla sense fregament és del tipus:

$$x(t) = Asin(\omega t + \varphi)$$

on $\omega$ és la freqüència angular i $\varphi$ és la fase inicial.

### E1.1
**Crea una funció `singen()`en Python per generar un sinus x(t) d'amplitud, durada, freqüència i fase inicial a determinar per l'usuari, i.e. com a paràmetres de la funció. Crea els senyals x1(t) i x2(t) amb els paràmetres indicats a continuació i fes dues gràfiques de x1 i x2 amb t a l'eix d'abscisses. Explica les diferències que veus entre les dues gràfiques.**

**[ENG] Create a `singen()` function that generates a sinusoid x(t) where amplitude, duration, frequency and phase are user inputs. Then run your function with the indicated inputs below and explain the differences between both plots.**

Per x1(t):

- *t* de 0 a 5 segons 
- passos de 0.01
- *f=1 Hz* 
- *A=1* 
- $\varphi$=0

Per x2(t):

- *t* de 0 a 5 segons
- passos de 0.08
- *f=10 Hz*
- *A=1*
- $\varphi$=0.

```python
# Completeu la funció, executeu-la per x1(t) i x2(t) amb els paràmetres indicats, i genereu els plots indicats

def singen(amp, tt, freq, phi):

    omega = 2*np.pi*freq
    sinusoid = amp*np.sin(omega*tt + phi)
    return sinusoid

fs = 16000.0
t_interval = np.arange(0, 5, 1/fs)
x1 = singen(1, t_interval, 1, 0)
x2 = singen(1, t_interval, 10, 0)

fig, (ax1, ax2) = plt.subplots(2)
ax1.plot(t_interval, x1,'b', label = 'x1')
ax1.set_title('x1(t)', fontsize=14, fontweight='bold')
ax2.plot(t_interval, x2,'g', label = 'x2')
ax2.set_title('x2(t)', fontsize=14, fontweight='bold')

ax1.label_outer()
ax2.label_outer()
ax2.set_xlabel('time [s]', fontsize=12)
fig.show()
```

The parameter that differentiates both plots is the frequency. Numerically, we know that f2 is 10 times f1, so it makes sense that when we observe the plots we see how for every cycle of x1, x2 completes 10 cycles. Also notice how when the frequency is 10 times greater, the period is 10 times smaller, which means that a cycle will be completed in much less time.

### E1.2
**Crea una nova funció `singen2()` que sigui com la de l'apartat E1.1 però modifica el codi per sumar un valor fix (p. ex. 0.5 o 0.8) a la variable temporal *t*. Utilitzant els mateixos paràmetres que a l'apartat E1.1, crea els senyals x3(t) i x4(t). Un cop fet això, representa gràficament x1 i x3 en un mateix _plot_ (colors diferents) i x2 i x4 en un altre _plot_ (també amb colors diferents). A continuació, explica què ha passat.**

**[ENG] Now create a `singen2()` that is similar to `singen()` but has an additional input that is used to sum a paticular fix value (ex. 0.5 or 0.8) to the temporal variable *t*. By summing a different value to x1 and x2 from E1.1, you will obtain x3 and x4. Plot x1 and x3 together and x2 and x4 together and explain what has happened.**


```python
# Completeu la funció, executeu-la per x3(t) i x3(t) amb els paràmetres indicats, i genereu els plots indicats

def singen2(amp, tt, freq, phi, value):
    tt = tt + value
    omega = 2*np.pi*freq
    sinusoid = amp*np.sin(omega*tt + phi)
    return sinusoid

value = 0.68
fs = 16000.0
t_interval = np.arange(0, 5, 1/fs)
x3 = singen2(1, t_interval, 1, 0, value)
x4 = singen2(1, t_interval, 10, 0, value)

fig, (ax1, ax2) = plt.subplots(2, figsize = (15,10))
ax1.plot(t_interval, x3,'b', label = 'x3')
ax1.plot(t_interval, x1,'y', label = 'x1')
ax1.legend()
ax1.set_title('x1(t) & x3(t)', fontsize=14, fontweight='bold')
ax2.plot(t_interval, x4,'r', label = 'x4')
ax2.plot(t_interval, x2,'g', label = 'x2')
ax2.legend()
ax2.set_title('x2(t) & x4(t)', fontsize=14, fontweight='bold')

ax1.label_outer()
ax2.label_outer()
ax2.set_xlabel('time [s]', fontsize=12)

fig.show()
```

Here, we can see how the original signal has been shifted. As the value has been added, the signal has been moved towards the left, that is, the signal has been advanced. Notice how if you set the value to a one-decimal float (0.3, 0.8...), x2 and x4 would still coincide in time because it would be a muktiple value of their period, so we need to add a second decimal to see a change. The same would happen with x1 and x3 if we set an integer value.

### E1.3
**Repeteix el mateix que a l'apartat E1.2 però en aquest cas modifica el codi per restar un valor fix a la variable temporal.
```python
# Completeu la funció, executeu-la per x5 i x6 i genereu els plots indicats

def singen3(amp, tt, freq, phi, value):
    tt = tt - value
    omega = 2*np.pi*freq
    sinusoid = amp*np.sin(omega*tt + phi)
    return sinusoid

value = 0.25
fs = 16000.0
t_interval = np.arange(0, 5, 1/fs)
x5 = singen3(1, t_interval, 1, 0, value)
x6 = singen3(1, t_interval, 10, 0, value)

fig, (ax1, ax2) = plt.subplots(2, figsize = (15,10))
ax1.plot(t_interval, x5,'b', label = 'x5')
ax1.plot(t_interval, x1,'y', label = 'x1')
ax1.plot(t_interval, x3,'c', label = 'x3')
ax1.legend()
ax1.set_title('x1(t), x3(t) & x5(t)', fontsize=14, fontweight='bold')

ax2.plot(t_interval, x6,'r', label = 'x6')
ax2.plot(t_interval, x2,'g', label = 'x2')
ax2.plot(t_interval, x4,'m', label = 'x4')
ax2.set_title('x2(t), x4(t) & x6(t)', fontsize=14, fontweight='bold')
ax2.legend()

ax1.label_outer()
ax2.label_outer()
ax2.set_xlabel('time [s]', fontsize=12)

fig.show()
```

## Exercici 2: Moviment harmònic esmorteït (3 punts)


Ara que ja hem vist que el desplaçament en un moviment harmònic simple segueix l'equació de la forma: 

$$x(t) = Asin(\omega t + \varphi)$$


i que l'energia del sistema no depèn de l'instant de temps i es calcula amb les següents equacions:


$$E_t = E_p + E_c$$
 
$$E_p(t) = \frac{1}{2}KA^2sin^2(\omega_0t+\varphi)$$

$$E_c(t)= \frac{1}{2}mA^2\omega_0^2cos^2(\omega_0t + \varphi)$$

$$E_t = \frac{1}{2}KA^2 = \frac{1}{2}mU^2$$ on $U=A\omega_0$ i $K=m\omega_0^2$.


Passem a estudiar el comportament del desplaçament i l'energia del moviment harmònic esmorteït, on el moviment disminueix la seva amplitud fins a aturar-se. En aquest cas tenim alguns paràmetres nous: $\lambda$ és el coeficient de fricció, $\tau$ és el temps de relaxació del sistema, i $\lambda$, que es relaciona de la següent manera: $\tau = \frac{1}{\lambda}$ (a vegades també $\tau = \frac{1}{2\lambda}$).

Amb aquests paràmetres, l'**equació del moviment** queda:

$$x(t) = Ae^{-\gamma t}sin(\omega t + \varphi)$$

on $\omega = \sqrt{\omega_0^2 - \gamma^2}$

Que també es pot escriure:

$$x(t) = Ae^{-\frac{t}{2\tau}}sin(\omega t + \varphi)$$

on $\omega = \sqrt{\omega_0^2 - \frac{1}{4\tau^2}}$

També recordem que les **energies** es calculen de la següent manera:

$$E_p(t) = \frac{1}{2}KA^2e^\frac{-t}{\tau}sin^2(\omega t + \varphi)$$

$$E_c(t) = \frac{1}{2}mA^2\omega^2e^\frac{-t}{\tau}cos^2(\omega t + \varphi)$$

### E2.1
**Crea tres funcions: `mov_esm()`, `potencial()`, `cinetica()` que implementin les seves respectives equacions. Els paràmetres d'entrada de les funcions han de ser totes les variables que no siguin fixes. Seguiu el model de l'exercici anterior.**

**[ENG] Create three functions: `mov_esm()`, `potencial()`, `cinetica()` that implement the respective equations. All the variables should be function inputs.**

### HINT: When gamma is greater than omega, the operation inside the square root of the damped omega is negative and then the square root will fail.
To solve this issue you may do `np.real(np.sqrt((a - b) + 0j))`

```python
# codi aquí
def mov_esm(amp, tt, gamma, omega, phi):
    x_t = amp*np.exp(-gamma*tt)*np.sin(omega*tt + phi)
    return x_t

def potencial(K, amp, tt, tau, omega,  phi):
    en = (K*(amp**2)*np.exp(-tt/tau)*(np.sin(omega*tt + phi))**2)/2
    return en

def cinetica(m, amp, tt, tau, omega,  phi):
    en = (m*(amp**2)*(omega**2)*np.exp(-tt/tau)*(np.cos(omega*tt + phi))**2)/2
    return en
```

**[ENG] Generate a function `equOna()` that has input parameters $A$, $\lambda$ i $v_p$. This function does not return anything, but generates a moving plot to represent a wave moving along the time. The moving space is 1m. Recommended steps to follow:**

```python
from IPython.display import display, clear_output
```

```python
# Codi aquí
def equOna(A, lamb, vp):

    # Calcular parametres
    freq = vp/lamb
    per = 1/freq
    k = 2*np.pi/lamb
    omega = 2*np.pi/per

    # Calcular vectors de temps i espai
    espai = np.arange(0, 1, 0.01)
    temps = np.arange(0, 1, 1/25)
    
    # Generar plot (podeu canviar-ho si ho considereu oportú)
    fig = plt.figure(figsize=(10,5))  # Creem una figura on anirem actualitzant el plot
    ax = fig.add_subplot(1,1,1)  # Declarem els eixos del plot per poder especificar-los després
    ax.set_xlim(0,1)
    ax.set_ylim(-A-0.2, A+0.2)

    # Loop sobre el vector de temps generant els plots de la ona
    for el in temps:
        y_x_t = A*np.sin(k*espai - omega*el)
        plt.plot(y_x_t)
        plt.xlabel('espai [cm]', fontsize=12)
        plt.pause(0.2)
        clear_output(wait= True)
    
    print(freq,per,k,omega)
```
